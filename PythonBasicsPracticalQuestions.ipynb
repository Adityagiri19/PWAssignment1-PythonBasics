```python
#Write a program to demonstrate the use of logical operators

def demonstrate_logical_operators(a, b):
    print(f"Values: a = {a}, b = {b}")

    # Logical AND
    print("\nLogical AND (a > 0 and b > 0):")
    print(f"a > 0 and b > 0: {a > 0 and b > 0}")

    # Logical OR
    print("\nLogical OR (a > 0 or b > 0):")
    print(f"a > 0 or b > 0: {a > 0 or b > 0}")

    # Logical NOT
    print("\nLogical NOT (not a > 0):")
    print(f"not a > 0: {not a > 0}")

# Test the function with different inputs
print("Example 1:")
demonstrate_logical_operators(5, -3)

print("\nExample 2:")
demonstrate_logical_operators(-2, 4)

print("\nExample 3:")
demonstrate_logical_operators(0, 0)
```

```python
#Write a Python program to convert user input from string to integer, float, and boolean types
user_input = input("Enter a value: ")

# Convert to integer
try:
    int_value = int(user_input)
    print(f"Integer conversion: {int_value}")
except ValueError:
    print("Cannot convert to an integer.")

# Convert to float
try:
    float_value = float(user_input)
    print(f"Float conversion: {float_value}")
except ValueError:
    print("Cannot convert to a float.")

# Convert to boolean
bool_value = bool(user_input)  # Converts any non-empty string to True
print(f"Boolean conversion: {bool_value}")

```

```python
even_sum = 0

for number in range(1, 51):
    if number % 2 == 0:
        even_sum += number

# Print the result
print("The sum of all even numbers between 1 and 50 is:", even_sum)

```

```python
#Write a program to reverse a string using a while loop
# Take input from the user
original_string = input("Enter a string: ")

# Initialize variables
reversed_string = ""
index = len(original_string) - 1

# Reverse the string using a while loop
while index >= 0:
    reversed_string += original_string[index]
    index -= 1  #

# Print the reversed string
print("Reversed string:", reversed_string)

```

```python
#Write a Python program to calculate the factorial of a number provided by the user using a while looP

number = int(input("Enter a positive integer: "))

if number < 0:
    print("Factorial is not defined for negative numbers.")
else:
    factorial = 1
    i = number

    # Calculate factorial using a while loop
    while i > 0:
        factorial *= i  # Multiply current value
        i -= 1  # Decrement the value of i

    # Print the result
    print(f"The factorial of {number} is: {factorial}")
```
