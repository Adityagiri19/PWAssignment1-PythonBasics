#Python Basics Questions
1) What is Python, and why is it popular?

 -> Python is a high-level, general-purpose programming language known for its simplicity, readability, and versatility. It was created by Guido van Rossum in 1991 and has become one of the most popular programming languages in the world.

  -> Python syntax is clean and easy to understand, making it an excellent choice for beginners. It reads like plain English, reducing the learning curve.

2) What is an interpreter in Python

-> An interpreter in Python is a program that executes Python code line by line. It takes the human-readable Python code, translates it into machine code, and runs it directly without needing to compile the entire program beforehand.

3) What are pre-defined keywords in Python

-> Pre-defined keywords in Python are reserved words that have specific meanings and purposes. These keywords are part of the language's syntax and cannot be used as variable, function, or identifier names.

4) Can keywords be used as variable names

-> No, keywords cannot be used as variable names in Python. Since keywords are reserved for specific functionalities in the language, using them as variable names would cause a syntax error.

5) What is mutability in Python

-> Mutability in Python refers to whether an objects value can be changed after it is created.
Mutable Objects in python are list, dict, set, bytearray.

6) Why are lists mutable, but tuples are immutable

-> Lists are mutable because they are designed for dynamic data manipulation, allowing in-place changes like adding, removing, or updating elements.

Tuples are immutable because they are intended for fixed data, ensuring integrity and enabling features like being hashable for use as dictionary keys or set elements.

7)  What is the difference between “==” and “is” operators in Python

-> The == operator checks if the values of two objects are equal, while the is operator checks if two objects refer to the same memory location

8)  What are logical operators in Python

-> Logical operators in Python are used to combine conditional statements and perform logical operations. These operators evaluate expressions and return a Boolean value (True or False).

9) What is type casting in Python
-> Type casting in Python is the process of converting one data type to another. There are two types of type casting:

Implicit Type Casting: Python automatically converts one data type to another, usually when performing operations between different types (e.g., an integer and a float).

Example:
x = 10     # int
y = 2.5    # float
result = x + y  # Python converts 'x' to float
print(result)  # Output: 12.5

Explicit Type Casting: The programmer manually converts a data type to another using functions like int(), float(), str(), etc.

Example:
x = 10.5
y = int(x)  # Converts float to int
print(y)  # Output: 10

10) What is the difference between implicit and explicit type casting

 -> Implicit casting is automatic, while explicit casting requires manual conversion by the programmer.

11)  What is the purpose of conditional statements in Python

-> The purpose of conditional statements in Python is to enable the program to make decisions based on certain conditions or criteria. These statements allow the program to execute different blocks of code depending on whether a specific condition is True or False.

12) How does the elif statement work

-> The elif short for "else if" statement in Python is used to check multiple conditions in a sequence. It allows you to test more than one condition in an if-else chain. If the condition in the if statement is False, Python will check the conditions in each elif block, one by one, in the order they appear. If an elif condition is True, the corresponding block of code is executed, and the rest of the conditions are not checked.

13) What is the difference between for and while loops

-> for loop is typically used when you know how many times you need to repeat something (fixed iteration).

while loop is used when the iteration depends on a condition that may change, and you dont know the number of iterations beforehand.


14) Describe a scenario where a while loop is more suitable than a for loop.

-> A while loop is more suitable when you dont know how many times a condition will hold true, and you want to keep repeating something until that condition becomes False.
